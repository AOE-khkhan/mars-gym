```python
_g = df.sort_values('pos_item_idx')\
                .groupby(['timestamp', 'step', 'session_idx'])



df_group                  = _g.agg({'impressions': 'first'})
df_group['list_item_idx'] = _g['item_idx'].apply(list)
df_group['pos_item_idx']  = _g['pos_item_idx'].apply(list)
df_group['clicked']       = _g['clicked'].apply(list)
df_group['item_idx']      = df_group.apply(lambda row: int(np.max(np.array(row['clicked'])*np.array(row['list_item_idx']))), axis=1)


df_group = df_group.reset_index()

df_group.head()
```

```python
_g = df.sort_values('pos_item_idx')\
                .groupby(['timestamp', 'step', 'user_idx', 'session_idx'])



df_group                  = _g.agg({'impressions': 'first'})
df_group['list_item_idx'] = _g['item_idx'].apply(list)
df_group['pos_item_idx']  = _g['pos_item_idx'].apply(list)
df_group['clicked']       = _g['clicked'].apply(list)
df_group['item_idx']      = df_group.apply(lambda row: int(np.max(np.array(row['clicked'])*np.array(row['list_item_idx']))), axis=1)


df_group = df_group.reset_index()

df_group.head()
```

```python
def sort_and_bin(row):
    list_sorted, score = zip(*sorted(zip(row['list_item_idx'], row['clicked']), key = lambda x: x[1], reverse=True))
    list_sorted = (np.array(list_sorted) == row['item_idx']).astype(int)
    
    return list(list_sorted)
   
df['sorted'] = df_group.apply(sort_and_bin, axis=1)
df.head()
```
